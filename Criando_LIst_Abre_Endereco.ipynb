# Abrindo as Base de Dados

```python
df_Abrev_Enderecos=pd.read_csv("Dados/Enderecos/List_Abrev_Ender_Correios_Sites.csv",encoding='utf-8')
```

```python
print ("Qtd. de Abrev.: "+str(len(df_Abrev_Enderecos)))
df_Abrev_Enderecos.head()
```

```python
df_Abrev_Enderecos_Comb=pd.DataFrame(columns=["Abrev_Enderecos_Comb","Nome_Abrev_Enderecos_Comb"])
cont=0
for Nome_Abrev in df_Abrev_Enderecos.Nome_Abrev_Enderecos:
    for Posc_2_Acima in range(2,len(Nome_Abrev)+1):
        df_Abrev_Enderecos_Comb.set_value(cont,"Abrev_Enderecos_Comb",Nome_Abrev[:Posc_2_Acima])
        df_Abrev_Enderecos_Comb.set_value(cont,"Nome_Abrev_Enderecos_Comb",Nome_Abrev)
        cont +=1
                                                                                            
        
```

```python
print ("Qtd. de Abrev. Combinadas: "+str(len(df_Abrev_Enderecos_Comb)))
df_Abrev_Enderecos_Comb_Sem_Supl=df_Abrev_Enderecos_Comb.drop_duplicates("Abrev_Enderecos_Comb")
print ("Qtd. de Abrev. Combinadas sem Duplicar: "+str(len(df_Abrev_Enderecos_Comb_Sem_Supl)))
```
